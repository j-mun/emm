# EMM custom material database

A custom material database can be used by `dat=` argument.

Additionally, this scheme can be used to manage materials in different projects.

```python
# import packages
import emm
```

#### Using custom material database

```python
# list of materials in the custom material database `cust_dat`
emm.avail(dat='cust_dat')

# list models of the material `mat1`
emm.avail('mat1', dat='cust_dat')
```

#### Using project-based material database

```python
# list of projects
emm.avail(dat='proj_dat')

# list materials under project 1
emm.avail('proj1', dat='proj_dat')

# list materials under project 2
emm.avail('proj2', dat='proj_dat')
```
