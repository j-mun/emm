# EMM Demo

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
import emm
```

```python
# check available materials
emm.avail()
```

```python
# check available model for a material
emm.avail('Ag')
```

```python
# import raw data
f,e = emm.read(
  'Ge/Aspnes',    # material/model
  unit='m',       # frequency/wavelength unit
  parm='e'        # parameter: e or n
)
plt.figure(figsize=[4,3.4])
plt.plot(f*1e9, np.real(e), '.-', label='Re(ε)')
plt.plot(f*1e9, np.imag(e), '.-', label='Im(ε)')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Relative permittivity (a.u.)')
plt.legend()
plt.show()
```

```python
# import interpolated data
lambda0 = np.linspace(400e-9, 900e-9, 101)
n = emm.load(
  'Ag/Johnson', # material/model
  lambda0,      # frequency/wavelength
  unit='m',     # frequency/wavelength unit
  parm='n'      # parameter: e or n
)
raw_w, raw_n = emm.read('Ag/Johnson', unit='m', parm='n')

plt.figure(figsize=[4,3.4])
plt.plot(lambda0*1e9, np.real(n), '-', label='Re(n)')
plt.plot(lambda0*1e9, np.imag(n), '-', label='Im(n)')
plt.plot(raw_w*1e9, np.real(raw_n), '.', label='Re(n), raw')
plt.plot(raw_w*1e9, np.imag(raw_n), '.', label='Im(n), raw')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Refractive index (a.u.)')
plt.legend()
plt.xlim(400, 900)
plt.ylim(-1, 7)
plt.show()
```

```python
# import interpolated data (cubic interp)
lambda0 = np.linspace(400e-9, 900e-9, 101)
n = emm.load('Ag/Johnson', lambda0, unit='m', parm='n', interp='cubic')
raw_w, raw_n = emm.read('Ag/Johnson', unit='m', parm='n')

plt.figure(figsize=[4,3.4])
plt.plot(lambda0*1e9, np.real(n), '-', label='Re(n)')
plt.plot(lambda0*1e9, np.imag(n), '-', label='Im(n)')
plt.plot(raw_w*1e9, np.real(raw_n), '.', label='Re(n), raw')
plt.plot(raw_w*1e9, np.imag(raw_n), '.', label='Im(n), raw')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Refractive index (a.u.)')
plt.legend()
plt.xlim(400, 900)
plt.ylim(-1, 7)
plt.show()
```
